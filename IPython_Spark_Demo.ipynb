## Creating a new Spark cell

1. Create a new cell above a code cell.
2. Set the new cell's type to "Spark" using the cell type selector in the IPython toolbar.
3. Type in the function description in the form `name type variant`, for example
    - `Foo unblocked 1` (becomes Foo_unb_var1)
    
    The type can be `unblocked`, `blocked`, or `recursive`. The variant can be an integer
    from `1` to `20`, or `none`.
4. Execute the cell (`SHIFT-ENTER`). If you made an error in the function description, it
    will  be displayed, and you can re-edit the cell (by double-clicking on it).

Try it now: select the next cell, choose "Insert Cell Above" from the toolbar, and change the type to Spark. Then type `Demo unblocked 1` and execute the Spark cell.

## Using the Spark menu

The spark menu works much like the one on the Spark website. When you change any of the selections, it regenerates the function and writes it into the following code cell. There are three differences:

1. You can change the parameters in the menu without worrying about overwriting any changes you make in the function; whatever you type in between the generated sections will be preserved.
2. The parts of the code generated by Spark are read-only, so you can't accidentally mess them up.
3. You change the function name by re-editing the Spark menu cell.
